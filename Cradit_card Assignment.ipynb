```python
columns_numeric =['LIMIT_BAL','AGE','BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6','PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6','QUALIFICATION','MARRIED']
columns_categorical=['PAY_0', 'GENDER', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
```

```python
# imputation
from sklearn.impute import SimpleImputer
imputer_num = SimpleImputer(strategy='median')
imputer_num.fit(df_train[columns_numeric])
imputer_cat = SimpleImputer(strategy='most_frequent')
imputer_cat.fit(df_train[columns_categorical])
```

```python
df_train[columns_numeric] = imputer_num.transform(df_train[columns_numeric])
df_test[columns_numeric] = imputer_num.transform(df_test[columns_numeric])

df_train[columns_categorical] = imputer_cat.transform(df_train[columns_categorical])
df_test[columns_categorical] = imputer_cat.transform(df_test[columns_categorical])
```

```python
df_train.isna().sum(),df_test.isna().sum()
```

```python
#scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df_train[columns_numeric])
df_train[columns_numeric] = scaler.transform(df_train[columns_numeric])
```

```python
df_train[columns_numeric]
```

```python
# encoding
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False,handle_unknown='ignore')
ohe.fit(df_train[columns_categorical])
```

```python
columns_final =list(ohe.get_feature_names())
print(columns_final)
```

```python
df_train[columns_final]=ohe.transform(df_train[columns_categorical])
df_test[columns_final]=ohe.transform(df_test[columns_categorical])

x_train = df_train[columns_numeric+columns_final]
x_test = df_test[columns_numeric+columns_final]
```

```python
# model build
from sklearn.linear_model import LogisticRegression
```

```python
mLogReg_final = LogisticRegression(random_state=19)
mLogReg_final.fit((df_train[columns_numeric+columns_final]), df_train['defaulter'])
```

```python
mLogReg_final.score((df_train[columns_numeric+columns_final]), df_train['defaulter'])
```

```python
yp =  mLogReg_final.predict(x_test)
```

```python
df_test['defaulter']=yp
```

```python
df_test[['Id','defaulter']].to_csv('Cradit_card.csv',index=False)
```
